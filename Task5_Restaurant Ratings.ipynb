## Analyze the distribution of aggregate  ratings and determine the most common rating range.

```python
rating_col = 'Aggregate rating'
```

```python
df[rating_col] = pd.to_numeric(df[rating_col], errors='coerce')
```

```python
df = df.dropna(subset=[rating_col])
```

```python
rating_bins = [0, 1, 2, 3, 4, 5.1]
rating_labels = ['0–1', '1–2', '2–3', '3–4', '4–5']
df['Rating Range'] = pd.cut(df[rating_col], bins=rating_bins, labels=rating_labels, include_lowest=True)
```

```python
rating_distribution = df['Rating Range'].value_counts().sort_index()
```

```python
print("\nDistribution of Aggregate Ratings:")
print(rating_distribution)
```

```python
most_common_range = rating_distribution.idxmax()
print(f"\nMost Common Rating Range: {most_common_range}")
```

```python
plt.figure(figsize=(8, 5))
rating_distribution.plot(kind='bar', color='dodgerblue')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating Range")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Calculate the average number of votes  received by restaurants.

```python
votes_col = 'Votes'
```

```python
df[votes_col] = pd.to_numeric(df[votes_col], errors='coerce')
```

```python
df_votes = df.dropna(subset=[votes_col])
```

```python
average_votes = df_votes[votes_col].mean()
print(f"\nAverage Number of Votes Received by Restaurants: {average_votes:.2f}")
```
