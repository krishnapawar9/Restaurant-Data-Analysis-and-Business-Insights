##  Plot the locations of restaurants on a map using longitude and latitude coordinates.

##  Identify any patterns or clusters of restaurants in specific areas.

```python
lat_col = 'Latitude'
lon_col = 'Longitude'
```

```python
df[lat_col] = pd.to_numeric(df[lat_col], errors='coerce')
df[lon_col] = pd.to_numeric(df[lon_col], errors='coerce')
df_geo = df.dropna(subset=[lat_col, lon_col])
```

```python
sample_size = 1000
```

```python
df_sample = df_geo.sample(n=min(sample_size, len(df_geo)), random_state=42)
print(f"\nTotal geolocated restaurants in sample: {len(df_sample)}")
```

```python
map_center = [df_sample[lat_col].mean(), df_sample[lon_col].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=11)
```

```python
for _, row in df_sample.iterrows():
    folium.CircleMarker(
        location=[row[lat_col], row[lon_col]],
        radius=2,
        color='blue',
        fill=True,
        fill_opacity=0.6
    ).add_to(restaurant_map)
```

```python
restaurant_map.save("restaurant_locations_map.html")
print("\nMap with restaurant markers saved as 'restaurant_locations_map.html'")
```

```python
heatmap = folium.Map(location=map_center, zoom_start=11)
heat_data = [[row[lat_col], row[lon_col]] for _, row in df_sample.iterrows()]
```

```python
HeatMap(heat_data, radius=10).add_to(heatmap)
heatmap.save("restaurant_heatmap.html")
print("Heatmap saved as 'restaurant_heatmap.html'")
```

```python
lat_col = 'Latitude'
lon_col = 'Longitude'
city_col = 'City'
cuisine_col = 'Cuisines'
```

#### Filter by City and/or Cuisine

```python
selected_city = "New Delhi"
selected_cuisine = "North Indian"

df_filtered = df.copy()
if selected_city:
    df_filtered = df_filtered[df_filtered[city_col].str.strip().str.lower() == selected_city.lower()]
if selected_cuisine:
    df_filtered = df_filtered[df_filtered[cuisine_col].str.contains(selected_cuisine, case=False, na=False)]

print(f"\nRestaurants matching filter: {len(df_filtered)}")
```

#### Plot Marker Map

```python
sample_size = 1000
df_sample = df_filtered.sample(n=min(sample_size, len(df_filtered)), random_state=42)

map_center = [df_sample[lat_col].mean(), df_sample[lon_col].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

for _, row in df_sample.iterrows():
    tooltip = f"{row.get('Restaurant Name', 'Restaurant')} ({row.get(cuisine_col, '')})"
    folium.CircleMarker(
        location=[row[lat_col], row[lon_col]],
        radius=3,
        color='blue',
        fill=True,
        fill_opacity=0.5,
        tooltip=tooltip
    ).add_to(restaurant_map)

restaurant_map.save("filtered_restaurant_map.html")
print("Filtered map saved as 'filtered_restaurant_map.html'")
```

```python
heatmap = folium.Map(location=map_center, zoom_start=12)
heat_data = [[row[lat_col], row[lon_col]] for _, row in df_sample.iterrows()]

HeatMap(heat_data, radius=10).add_to(heatmap)
heatmap.save("filtered_restaurant_heatmap.html")
print("Filtered heatmap saved as 'filtered_restaurant_heatmap.html'")
```
