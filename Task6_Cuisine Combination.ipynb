## Identify the most common combinations of cuisines in the dataset.

```python
cuisine_col = 'Cuisines'
rating_col = 'Aggregate rating'
```

```python
df = df.dropna(subset=[cuisine_col])
```

```python
df[cuisine_col] = df[cuisine_col].astype(str).str.strip()
```

```python
cuisine_combos = df[cuisine_col].value_counts().head(10)
```

```python
print("\nTop 10 Most Common Cuisine Combinations:")
print(cuisine_combos)
```

```python
plt.figure(figsize=(10, 5))
cuisine_combos.plot(kind='barh', color='coral')
plt.title("Top 10 Most Common Cuisine Combinations")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Combination")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Determine if certain cuisine combinations tend to have higher ratings.

```python
df[rating_col] = pd.to_numeric(df[rating_col], errors='coerce')
```

```python
df_rated = df.dropna(subset=[rating_col])
```

```python
avg_rating_by_combo = df_rated.groupby(cuisine_col)[rating_col].mean()
```

```python
min_count_threshold = 5
combo_counts = df[cuisine_col].value_counts()
valid_combos = combo_counts[combo_counts >= min_count_threshold].index
```

```python
top_rated_combos = avg_rating_by_combo[avg_rating_by_combo.index.isin(valid_combos)].sort_values(ascending=False).head(10)
print("\nTop 10 Cuisine Combinations by Average Rating (min 5 restaurants):")
print(top_rated_combos)
```

```python
plt.figure(figsize=(10, 5))
top_rated_combos.plot(kind='barh', color='seagreen')
plt.title("Top 10 Cuisine Combinations by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Cuisine Combination")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
