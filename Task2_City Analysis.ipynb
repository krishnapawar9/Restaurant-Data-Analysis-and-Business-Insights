## Identify the city with the highest number of restaurants in the dataset.

```python
city_col = 'City'
```

```python
city_counts = df[city_col].value_counts()
```

```python
print("\nCity with the Highest Number of Restaurants:")
print(city_counts.head(1))
```

```python
plt.figure(figsize=(10, 5))
city_counts.head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Calculate the average rating for restaurants in each city.

```python
rating_col = 'Aggregate rating'
```

```python
df[rating_col] = pd.to_numeric(df[rating_col], errors='coerce')
```

```python
df_clean = df.dropna(subset=[rating_col])
```

```python
avg_ratings = df_clean.groupby(city_col)[rating_col].mean().sort_values(ascending=False)
```

```python
print("\nAverage Rating by City:")
print(avg_ratings.head(10))
```

```python
plt.figure(figsize=(10, 5))
avg_ratings.head(10).plot(kind='bar', color='green')
plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.ylim(0, 5)  # Ratings are usually between 0 and 5
plt.tight_layout()
plt.show()
```
