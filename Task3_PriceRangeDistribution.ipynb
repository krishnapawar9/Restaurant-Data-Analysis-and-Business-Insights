##  Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
price_col = 'Price range'
```

```python
df = df.dropna(subset=[price_col])
```

```python
df[price_col] = pd.to_numeric(df[price_col], errors='coerce').astype('Int64')
```

```python
price_counts = df[price_col].value_counts().sort_index()
```

```python
plt.figure(figsize=(8, 5))
price_counts.plot(kind='bar', color='teal')
plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range (1 = Low, 4 = High)")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Calculate the percentage of restaurants in each price range category.

```python
total_restaurants = len(df)
```

```python
price_percentages = (price_counts / total_restaurants) * 100
```

```python
print("\nPercentage of Restaurants in Each Price Range:")
for price, percent in price_percentages.items():
    print(f"Price Range {price}: {percent:.2f}%")
```

```python
plt.figure(figsize=(7, 7))
plt.pie(
    price_percentages,
    labels=[f"Price {i}" for i in price_percentages.index],
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title("Percentage of Restaurants by Price Range")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
