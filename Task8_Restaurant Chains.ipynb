##  Identify if there are any restaurant chains  present in the dataset. 

```python
name_col = 'Restaurant Name'
chain_counts = df[name_col].value_counts()
chains = chain_counts[chain_counts > 1]

print(f"\nTotal restaurant chains found: {len(chains)}")
print("\nTop 10 restaurant chains by number of locations:")
print(chains.head(10))
```

## Analyze the ratings and popularity of  different restaurant chains.

```python
rating_col = 'Aggregate rating'
votes_col = 'Votes'
```

```python
df[rating_col] = pd.to_numeric(df[rating_col], errors='coerce')
df[votes_col] = pd.to_numeric(df[votes_col], errors='coerce')
```

```python
df_chains = df[df[name_col].isin(chains.index)]

chain_analysis = df_chains.groupby(name_col).agg({
    rating_col: 'mean',
    votes_col: 'sum',
    name_col: 'count'
}).rename(columns={
    rating_col: 'Average Rating',
    votes_col: 'Total Votes',
    name_col: 'Number of Locations'
}).sort_values(by='Number of Locations', ascending=False)
```

```python
print("\nTop 10 Restaurant Chains by Popularity and Rating:")
print(chain_analysis.head(10))
```

####  Top 10 Chains by Number of Locations

```python
top_by_locations = chain_analysis.head(10)

plt.figure(figsize=(10, 5))
top_by_locations['Number of Locations'].plot(kind='bar', color='mediumblue')
plt.title("Top 10 Restaurant Chains by Number of Locations")
plt.ylabel("Number of Locations")
plt.xlabel("Restaurant Chain")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Top 10 Chains by Average Rating

```python
popular_chains = chain_analysis[chain_analysis['Number of Locations'] >= 3]
top_by_rating = popular_chains.sort_values(by='Average Rating', ascending=False).head(10)

plt.figure(figsize=(10, 5))
top_by_rating['Average Rating'].plot(kind='bar', color='darkgreen')
plt.title("Top 10 Restaurant Chains by Average Rating")
plt.ylabel("Average Rating")
plt.xlabel("Restaurant Chain")
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```
