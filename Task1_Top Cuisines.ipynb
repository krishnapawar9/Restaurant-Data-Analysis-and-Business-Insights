```python
cuisines_col = 'Cuisines'
```

```python
df[cuisines_col] = df[cuisines_col].astype(str)
```

```python
df[cuisines_col] = df[cuisines_col].str.split(',')
```

```python
exploded = df.explode(cuisines_col)
```

```python
top_cuisines = exploded[cuisines_col].value_counts().head(3)
```

```python
print("Top 3 Cuisines:")
print(top_cuisines)
```

```python
total_restaurants = len(df)
```

```python
percentages = (top_cuisines / total_restaurants) * 100
```

```python
print("\nPercentage of Restaurants Serving Top Cuisines:")
for cuisine, percent in percentages.items():
    print(f"{cuisine}: {percent:.2f}%")
```

```python
# Bar chart of top 3 cuisines
plt.figure(figsize=(8, 5))
top_cuisines.plot(kind='bar', color='skyblue')
plt.title("Top 3 Cuisines by Number of Restaurants")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Pie chart showing percentage
plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', colors=['#66c2a5','#fc8d62','#8da0cb'])
plt.title("Percentage of Restaurants Serving Top 3 Cuisines")
plt.axis('equal')
plt.show()
```
